```python
n=2
w=[1/2]*n
a=0
b=10
x=[a,b]
s=0
for i in range(0, n):
    s +=w[i]*f(x[i])
int=(b-a)*s
print(int)
print("Error:", intstd-int)
```

## 7.

(a) 

$$m\ddot{x}_n+kx_n=0$$

C.F.
$$m\lambda^2+k=0$$
$$\lambda=\pm i\sqrt{\frac{k}{m}}$$
$$x_n=A\cos{\sqrt{\frac{k}{m}}t}+Bi\sin{\sqrt{\frac{k}{m}}t}$$

P.I.
$$x_n=a$$
$$\ddot{x}=0$$
$$ka=0$$
$$a=0$$

$$x_n=A\cos{\sqrt{\frac{k}{m}}t}+Bi\sin{\sqrt{\frac{k}{m}}t}$$

Solve it
$$\dot{x}_n=-A\sqrt{\frac{k}{m}}\sin{\sqrt{\frac{k}{m}}t}+Bi\sqrt{\frac{k}{m}}\cos{\sqrt{\frac{k}{m}}t}$$
$$\ddot{x}_n=-A\frac{k}{m}\cos{\sqrt{\frac{k}{m}}t}-Bi\frac{k}{m}\sin{\sqrt{\frac{k}{m}}t}$$

thus
$$B=0\text{, }A=0.01$$
$$x_n=0.01\cos{\sqrt{\frac{k}{m}}t}$$

(b)
$$x_{n+1}=\frac{\Delta t^2 F_n + (2m-\Delta t^2 k)x_n-mx_{n-1}}{m}$$

```python
import numpy as np
a=0
b=20
dt=0.01
m=1
k=40
f=0
t=np.arange(0,20,0.01)
x=[0]*len(t)
x[0]=0.01
x[1]=0.01
for i in range (2, len(t)):
    x[i]=(dt**2*f+(2*m-dt**2*k)*x[i-1]-m*x[i-2])/m
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(t, x, label="numerical")
plt.plot(t, 0.01*np.cos(t*(k/m)**0.5), label="analytical")
plt.xlabel('t')
plt.legend();
```
