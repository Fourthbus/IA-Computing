## Exercise 11.1

```python
# Vector_vector product
def dot(vec_a, vec_b):
    if len(vec_a) == len(vec_b):
        product = 0
        n=len(vec_a)
        for i in range (n):
            product += vec_a[i]*vec_b[i]
        return product
    else:
        raise ValueError('The length of two vectors are not the same.')
```

The cost of the operation `vec_a[i]*vec_b[i]` is $O(1)$.

The `for` loop repeats `n` times, so the total cost is $O(n)$.

```python
# Matrix-vector product
def prod(matx, vec):
    if len(matx[0,:])==len(vec):
        m = len(matx[:,0])
        n = len(matx[0,:])
        product = [0]*m
        for i in range(m):
            for j in range(n):
                product[i] += matx[i,j]*vec[j]
        return product
    else:
        raise ValueError('Dimensions are not the same.')
```

The cost of the operation `matx[i,j]*vec[j]` is $O(1)$.

The first `for` loop repeats `n` times, so the cost is $O(n)$.

The second `for` loop repeats `n` times again, so the total cost is $O(n^2)$.

```python
# Matrix-matrix product
def cross(a, b):
    if len(a[0,:])==len(b[:,0]):
        m = len(a[:,0])
        n = len(b[0,:])
        p = len(b[:,0])
        product = np.array([[0]*n]*m)
        for i in range(m):
            for j in range(n):
                for k in range(p):
                    product[i,j] += a[i,k]*b[k,j]
        return product
    else:
        raise ValueError('Dimensions are not the same.')
```

The cost of the operation `matx[i,j]*vec[j]` is $O(1)$.

The first `for` loop repeats `n` times, so the cost is $O(n)$.

The second `for` loop repeats `n` times, so the cost is $O(n^2)$.

The third `for` loop repeats `n` times again, so the total cost is $O(n^3)$.

## Exercise 11.2

```python
def fac (n):
    'compute factorial for recursion'
    if n == 1:
        return 1
    else:
        return (n) * fac (n-1)
    
times = []

N = np.arange(1,20)

for n in N:
    t = %timeit -q -o fac(n)
    times.append(t.best)
    
# Plot and label the time taken for linear search
plt.loglog(N, times, marker='o', label='linear search')
plt.xlabel('$n$')
plt.ylabel('$t$ (s)')

# Show a reference line of O(n)
plt.loglog(N, 1e-6*N, label='$O(n)$')

# Add legend
plt.legend(loc=0);
```

The cost of operation is $O(n)$
