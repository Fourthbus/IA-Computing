## Exercise 10.1
searching a long list of random numbers

```python
import numpy as np

def min_max1(x):
    
    minim = None
    maxim = None
    
    for val in x:
        if minim == maxim == None:
            minim = maxim = val
            
        elif val < minim:
            minim = val
        elif val > maxim:
            maxim = val
    
    return minim, maxim

num_list = np.random.rand(1000)

print(min_max1(num_list))
        
```

```python
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

def newton(f=None, df=None, x0=None, tol = 1e-6, max_it=50):
    "Returns the approximate root of function f using Newton's method"
    
    root_try = x0 
    
    for i in range(max_it):
        
        root_try = root_try - f(root_try)/df(root_try)
        
        if np.abs(f(root_try)) <= tol:
            return root_try
    raise ValueError("max iterations reached")

def f(x):
    
    return np.tan(x) - 2*x

def df(x):
    
    return 1/(np.cos(x))**2 - 2

roots = [0,0,0]
zerosar = np.zeros(3)

roots[0] = newton(f,df,-1,1e-6)
roots[1] = newton(f,df,0,1e-6)
roots[2] = newton(f,df,1,1e-6)

print("root1:",roots[0])
print("root2:",roots[1])
print("root3:",roots[2])

x = np.linspace(-2*np.pi, 2*np.pi, 800)
f_x = f(x)
df_x = df(x)

plt.plot(x,f_x,'b', label = '$tan(x) -2x$')
plt.plot(x, df_x,'g', label = '$d/dx(tan(x) -2x)$')
plt.plot(roots, zerosar,'ro',label = '$roots$')
plt.ylim(-5,5);
plt.xlim(-2,2);
plt.legend(loc=0);
```

```python
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline

def newton2(f=None, df=None, x0=None, tol = 1e-6, max_it=50,dx=0.001):
    "Returns the approximate root of function f using Newton's method"
    
    root_try = x0 
    
    for i in range(max_it):
        
        root_try = root_try - f(root_try)/df2(f, dx, root_try)
        
        if np.abs(f(root_try)) <= tol:
            return root_try
    raise ValueError("max iterations reached")

def f(x):
    
    return np.tan(x) - 2*x

def df2(f, dx, x):
    
    return (f(x + dx) - f(x - dx))/(2*dx)

```

```python
roots = [0,0,0]
zerosar = np.zeros(3)

roots[0] = newton2(f,df,-1,1e-6)
roots[1] = newton2(f,df,0,1e-6)
roots[2] = newton2(f,df,1,1e-6)

print("root1:",roots[0])
print("root2:",roots[1])
print("root3:",roots[2])
```
