## Exercise 11.1

### Vector Dot Product
Given that:
$$\begin{pmatrix}a\\b\\c\end{pmatrix} \cdot \begin{pmatrix}d\\e\\f\end{pmatrix} = ad+be+cf$$
For matrix length n, n arithmatic operations must take place.

Artithmatic is $O(1)$ thus vector dot product is $O(n)$ complexity in all cases.

### Matrix-vector Product
Given that:
$$\begin{pmatrix}a & b & c\\
                 d & e & f\\
                 g & h & i\end{pmatrix}\begin{pmatrix}j\\k\\l\end{pmatrix}=\begin{pmatrix}aj + bk + cl
                                                                                        \\dj + ek + fl
                                                                                        \\gj + hk + il
                                                                                        \end{pmatrix}$$
Then for a matrix $m\times n$ and vector $n$ the complexity is $O(mn)$ so for a square matrix $n$ is $O(n^2)$

### Matrix-matrix product
Given that $$\begin{pmatrix}a & b\\c & d\end{pmatrix}\begin{pmatrix}e & f\\g & h\end{pmatrix} = \begin{pmatrix}ae + bg & af + bh\\ce + dg & cf + dh\end{pmatrix}$$
Then for matrices $m\times n$ and $n\times p$ complexity is $O(mnp)$ so for two square matrices $n$ complexity is $O(n^3)$

## Exercise 11.2

given that the code to compute n! factorial is:

```python
def factorial(n):
    if n == 1:
        return 1
    else: 
        return n*factorial(n-1)
```

it must calculate ```n*factorial(n-1)``` which is two arithmatic operations. It must do this $n$ times. Arithmatic is O(1) (ususally). So complexity is O(n)

````python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math

%matplotlib inline
````

````python
def factorial(n):
    if n == 1:
        return 1
    else: 
        return n*factorial(n-1)

times = []

N = np.arange(1,20)

for n in N:
    t = %timeit -q -o factorial(n)
    times.append(t.best)
    
plt.loglog(N,times);
plt.loglog(N,N*1e-6) ;
````

````python
N = 2**np.arange(2,15)

times = []

for n in N:
    mat = np.random.rand(n, n)
    t = %timeit -q -o -r3 np.linalg.slogdet(mat)
    times.append(t.best)

plt.loglog(N,1e-6*N**2.373);
plt.loglog(N,times);
````
